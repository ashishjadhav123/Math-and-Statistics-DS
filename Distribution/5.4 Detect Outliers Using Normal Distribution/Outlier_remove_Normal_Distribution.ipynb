```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_input = pd.read_csv("heights_with_outliers.csv")
df_input.head()
```

```python
mean_height = df_input['height'].mean()
mean_height
```

```python
std_height = df_input['height'].std()
std_height
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_input['height'])

plt.axvline(mean_height, color='red', linestyle='dashed', linewidth=1, label='Mean')

plt.axvline(mean_height + 1 * std_height, color='green', linestyle='dashed', linewidth=1, label='Mean + 1*Std')
plt.axvline(mean_height - 1 * std_height, color='green', linestyle='dashed', linewidth=1, label='Mean - 1*Std')

plt.axvline(mean_height + 2 * std_height, color='blue', linestyle='dashed', linewidth=1, label='Mean + 2*Std')
plt.axvline(mean_height - 2 * std_height, color='blue', linestyle='dashed', linewidth=1, label='Mean - 2*Std')

plt.axvline(mean_height + 3 * std_height, color='orange', linestyle='dashed', linewidth=1, label='Mean + 3*Std')
plt.axvline(mean_height - 3 * std_height, color='orange', linestyle='dashed', linewidth=1, label='Mean - 3*Std')    

plt.title('Height Distribution with Outliers')
plt.xlabel('Height (inches)')

plt.show()
```

```python
lower_limit = mean_height - 3 * std_height
upper_limit = mean_height + 3 * std_height

print(f"Lower Limit: {lower_limit}")
print(f"Upper Limit: {upper_limit}")
```

```python
df_no_outliers = df_input[(df_input['height'] >= lower_limit) & (df_input['height'] <= upper_limit)]
df_no_outliers.shape
```
