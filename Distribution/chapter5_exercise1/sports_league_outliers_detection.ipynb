## Problem Statement


You are a data scientist working for a sports analytics company. The dataset, named **"sports_league_data.csv"** includes a variety of attributes for players in the league. The columns in the dataset are as follows:

- **player_id:** A unique identifier for each player.
- **age:** Age of the player.
- **position:** The position the player typically plays in (e.g., Forward, Midfielder, Defender, Goalkeeper).
- **team:** The team the player belongs to.
- **games_played:** The total number of games played by the player in the season.
- **player_rating:** An overall performance rating for the player for the season.

**Import Necessary Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Task1

1. Import the data from the "sports_league_data.csv" file.
2. display the number of rows and columns. 
3. Display the first few rows of the dataset to get an overview.


```python
df_sports = pd.read_csv('sports_league_data.csv')

# Display the first few rows of the dataset
print(df_sports.shape)

df_sports.head()

```

## Task2:

**Age Analysis:**
- Mean: Calculate the average age of players in the league.
- Standard Deviation Intervals:
    - Calculate the age range within one standard deviation of the mean.
    - Calculate the age range within two standard deviations of the mean.
    - Calculate the age range within three standard deviations of the mean.
  

```python
age_mean = df_sports['age'].mean()
print(f"Mean age: {age_mean}")

age_std = df_sports['age'].std()
print(f"Standard deviation of age: {age_std}")
```

```python
upeer_one = age_mean + age_std
print(f"Mean + 1 standard deviation: {upeer_one}")

lower_one = age_mean - age_std
print(f"Mean - 1 standard deviation: {lower_one}")

upper_two = age_mean + 2 * age_std
print(f"Mean + 2 standard deviations: {upper_two}")

lower_two = age_mean - 2 * age_std
print(f"Mean - 2 standard deviations: {lower_two}")

upper_three = age_mean + 3 * age_std
print(f"Mean + 3 standard deviations: {upper_three}")

lower_three = age_mean - 3 * age_std
print(f"Mean - 3 standard deviations: {lower_three}")

```

## Task3:

**Player Rating Analysis:**
- Mean: Determine the average player rating across the league.
- Standard Deviation Intervals:
    - Calculate the player rating range within one standard deviation of the mean.
    - Calculate the player rating range within two standard deviations of the mean.
    - Calculate the player rating range within three standard deviations of the mean.

```python
rating_mean = df_sports['player_rating'].mean()
print(f"Mean rating: {rating_mean}")

rating_std = df_sports['player_rating'].std()
print(f"Standard deviation of rating: {rating_std}")

upper_one_rating = rating_mean + rating_std
print(f"Mean + 1 standard deviation: {upper_one_rating}")

lower_one_rating = rating_mean - rating_std
print(f"Mean - 1 standard deviation: {lower_one_rating}")

upper_two_rating = rating_mean + 2 * rating_std
print(f"Mean + 2 standard deviations: {upper_two_rating}")

lower_two_rating = rating_mean - 2 * rating_std
print(f"Mean - 2 standard deviations: {lower_two_rating}")

upper_three_rating = rating_mean + 3 * rating_std
print(f"Mean + 3 standard deviations: {upper_three_rating}")

lower_three_rating = rating_mean - 3 * rating_std
print(f"Mean - 3 standard deviations: {lower_three_rating}")
```

## Task4:

Create histogram with Kernel Density Estimation (KDE) plots for both the 'age' and 'player_rating' columns from the dataset. While constructing these plots, ensure to visually represent the standard deviation intervals on the graph.


- **'age'**

```python
plt.figure(figsize=(12, 6))

sns.histplot(df_sports['age'], bins=30, kde=True, color='blue', label='Age')

plt.axvline(age_mean, color='green', linestyle='--', label='Mean Age')

plt.axvline(upeer_one, color='gray', linestyle='--', label='Mean + 1 Std Dev')
plt.axvline(lower_one, color='gray', linestyle='--', label='Mean - 1 Std Dev')

plt.axvline(upper_two, color='orange', linestyle='--', label='Mean + 2 Std Dev')
plt.axvline(lower_two, color='orange', linestyle='--', label='Mean - 2 Std Dev')

plt.axvline(upper_three, color='red', linestyle='--', label='Mean + 3 Std Dev')
plt.axvline(lower_three, color='red', linestyle='--', label='Mean - 3 Std Dev')

plt.title('Age Distribution with Mean and Standard Deviations')
plt.xlabel('Age')
plt.legend()
plt.show()
```

- **'player_rating'**

```python
plt.figure(figsize=(12, 6))
sns.histplot(df_sports['player_rating'], bins=30, kde=True, color='blue', label='Player Rating')

plt.axvline(rating_mean, color='green', linestyle='--', label='Mean Rating')

plt.axvline(upper_one_rating, color='gray', linestyle='--', label='Mean + 1 Std Dev')
plt.axvline(lower_one_rating, color='gray', linestyle='--', label='Mean - 1 Std Dev')

plt.axvline(upper_two_rating, color='orange', linestyle='--', label='Mean + 2 Std Dev')
plt.axvline(lower_two_rating, color='orange', linestyle='--', label='Mean - 2 Std Dev')

plt.axvline(upper_three_rating, color='red', linestyle='--', label='Mean + 3 Std Dev')
plt.axvline(lower_three_rating, color='red', linestyle='--', label='Mean - 3 Std Dev') 
 
plt.title('Player Rating Distribution with Mean and Standard Deviations')
plt.xlabel('Player Rating')
plt.legend()
plt.show()
```

## Task5

- Calculate the lower and upper boundaries for the 'Age' and 'Player_Rating' columns. 
- For each column, identify values that fall outside the calculated lower and upper boundaries. These values are considered outliers.

- **Display age outliers**

```python
loweer_age = lower_three
upper_age = upper_three

print(f"Lower age threshold for outliers: {loweer_age}")
print(f"Upper age threshold for outliers: {upper_age}")
```

- **Display player_rating outliers**

```python
lower_rating = lower_three_rating
upper_rating = upper_three_rating

print(f"Lower rating threshold for outliers: {lower_rating}")
print(f"Upper rating threshold for outliers: {upper_rating}")
```

## Task6

-  Remove outliers from both the 'age' and 'player_rating' columns and then visualize the adjusted data using histogram plots.


```python
df_age_sport = df_sports.loc[(df_sports['age'] >= loweer_age) & (df_sports['age'] <= upper_age)]
df_age_sport

df_rating_sport = df_sports.loc[(df_sports['player_rating'] >= lower_rating) & (df_sports['player_rating'] <= upper_rating)]
df_rating_sport 
```

- **plot the histograms**

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_age_sport['age'], kde=True)
plt.title('Age Distribution After Outlier Removal')
plt.xlabel('Age')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df_rating_sport['player_rating'], kde=True)
plt.title('Player Rating Distribution After Outlier Removal')
plt.xlabel('Player Rating')
plt.show()
```
