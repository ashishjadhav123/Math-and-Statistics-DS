```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_height = pd.read_csv(r"heights.csv")
df_height.head()
```

```python
df_height.shape
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_height['height'], kde=True, stat="density", bins=30)
plt.title('Height Distribution')
plt.xlabel('Height (inches)')
plt.ylabel('Density')
# plt.axvline(df_height['height'].mean(), color='red', linestyle='dashed', linewidth=1, label='Mean')
# plt.axvline(df_height['height'].median(), color='green', linestyle='dashed', linewidth=1, label='Median')
# plt.axvline(df_height['height'].mode()[0], color='blue', linestyle='dashed', linewidth=1, label='Mode')
plt.legend()   
plt.show()
```

```python
mean_height = df_height['height'].mean()
std_height = df_height['height'].std()
print(f"Mean Height: {mean_height:.2f} inches")
print(f"Standard Deviation of Height: {std_height:.2f} inches")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_height['height'], kde=True, stat="density", bins=30)
plt.title('Height Distribution')
plt.xlabel('Height (inches)')
plt.ylabel('Density')
plt.axvline(mean_height, color='Black', linestyle='dashed', linewidth=1, label='Mean')

plt.axvline(mean_height + std_height, color='red', linestyle='dashed', linewidth=1, label='STD1')
plt.axvline(mean_height - std_height, color='red', linestyle='dashed', linewidth=1, label='STD1')

plt.legend()   
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_height['height'], kde=True, stat="density", bins=30)
plt.title('Height Distribution')
plt.xlabel('Height (inches)')
plt.ylabel('Density')
plt.axvline(mean_height, color='Black', linestyle='dashed', linewidth=1, label='Mean')

plt.axvline(mean_height + std_height, color='red', linestyle='dashed', linewidth=1, label='STD1')
plt.axvline(mean_height - std_height, color='red', linestyle='dashed', linewidth=1, label='STD1')

plt.axvline(mean_height + 2*std_height, color='green', linestyle='dashed', linewidth=1, label='STD2')
plt.axvline(mean_height - 2*std_height, color='green', linestyle='dashed', linewidth=1, label='STD2')

plt.axvline(mean_height + 3*std_height, color='blue', linestyle='dashed', linewidth=1, label='STD3')
plt.axvline(mean_height - 3*std_height, color='blue', linestyle='dashed', linewidth=1, label='STD3')

plt.legend()   
plt.show()
```

** Now let's calculate persentage value comes unser STD1, STD2 and STD3

```python
left1 = mean_height - std_height
right1 = mean_height + std_height
print(f"Left 1 STD: {left1:.2f} inches and Right 1 STD: {right1:.2f} inches")
```

```python
std1_per = df_height.loc[(df_height['height'] >= left1) & (df_height['height'] <= right1)].shape[0] / df_height.shape[0] * 100
print(f"Percentage of data within 1 standard deviation: {std1_per:.2f}%")
```

```python
left2 = mean_height - 2*std_height
right2 = mean_height + 2*std_height
print(f"Left 2 STD: {left2:.2f} inches and Right 2 STD: {right2:.2f} inches")
```

```python
std2_per = df_height.loc[(df_height['height'] >= left2) & (df_height['height'] <= right2)].shape[0] / df_height.shape[0] * 100
print(f"Percentage of data within 2 standard deviations: {std2_per:.2f}%")
```

```python
left3 = mean_height - 3*std_height
right3 = mean_height + 3*std_height
print(f"Left 3 STD: {left3:.2f} inches and Right 3 STD: {right3:.2f} inches")
```

```python
std3_per = df_height.loc[(df_height['height'] >= left3) & (df_height['height'] <= right3)].shape[0] / df_height.shape[0] * 100
print(f"Percentage of data within 3 standard deviations: {std3_per:.2f}%")
```
