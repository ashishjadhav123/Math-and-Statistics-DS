```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_input = pd.read_csv('correlation.csv')
df_input
```

```python
correlation =  df_input['Square Ft'].corr(df_input['Price (INR Cr)'])
print(f"Correlation between Square Ft and Price (INR Cr): {correlation:.2f}")
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df_input, x='Square Ft', y='Price (INR Cr)', color='blue', s=100, alpha=0.6, edgecolor='w')
plt.title('Scatter Plot of Square Ft vs Price (INR Cr)')
plt.xlabel('Square Ft') 
plt.ylabel('Price (INR Cr)')
plt.grid(True)
plt.show()
```
