# Nike Analysis 

```python
df_nike = df_input.loc[df_input['brand'] == 'Nike']
df_nike
```

```python
df_nike.describe()
```

```python
round(df_nike['sold_qty'].median())
```

```python
df_nike.isnull().sum()
```

```python
df_nike['sold_qty'].fillna(round(df_nike['sold_qty'].median()), inplace=True)
```

```python
df_nike.isnull().sum()
```

# Adidas Analysis

```python
df_adidas = df_input.loc[df_input['brand'] == 'Adidas']
df_adidas
```

```python
df_adidas.describe()
```

```python
median_adidas = round(df_adidas['sold_qty'].median())
median_adidas
```

```python
df_adidas.loc[df_adidas['sold_qty'] > median_adidas]
```

```python
quantile_val = round(df_adidas['sold_qty'].quantile(0.95))
quantile_val
```

```python
df_adidas.loc[df_adidas['sold_qty'] >  quantile_val]
```

* Remove Outlier from Addidas data frame

```python
df_adidas.loc[df_adidas['sold_qty'] > quantile_val, 'sold_qty'] = quantile_val
```

```python
df_adidas.loc[df_adidas['sold_qty'] >  quantile_val]
```

```python
df_adidas.plot(kind='bar', x='date', y='sold_qty', title='Adidas Shoes Sales Analysis', figsize=(10, 5))
```

```python
plt.figure(figsize=(15, 6))
plt.plot(df_nike['date'], df_nike['sold_qty'], color='blue', marker='o', label='Nike')
plt.plot(df_adidas['date'], df_adidas['sold_qty'], color='red', marker='o', label='Adidas')


# Customize plot
plt.title('Nike vs Adidas Shoe Sales in September')
plt.xlabel('Date')
plt.ylabel('Number of Shoes Sold')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```
